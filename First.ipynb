```python
display(combustiveis_df.loc[combustiveis_df['Estado - Sigla'] == "MG", ['Revenda', 'Municipio', 'Produto', 'Valor de Venda']])
```

```python
display(combustiveis_df.loc[combustiveis_df['Estado - Sigla'] == "MG", ['Revenda', 'Municipio', 'Produto', 'Valor de Venda']].max())
```

```python
etanol_bh_df = combustiveis_df.loc[(combustiveis_df['Produto'] == 'ETANOL') & (combustiveis_df['Municipio'] == 'BELO HORIZONTE'), ['Produto','Municipio', 'Nome da Rua', 'Bairro', 'Valor de Venda']].sort_values(by="Valor de Venda")
#etanol_bh_df.sort_values(by='Valor de Venda', inplace=True)
display(etanol_bh_df)
```

```python
etanol_bh_df.loc[etanol_bh_df['Bairro'] == 'CENTRO']
```

```python
#Média do valor de venda dos combustiveis
gas_df = combustiveis_df[['Produto', 'Estado - Sigla', 'Valor de Venda']].groupby(['Produto']).mean().round(2)
display(gas_df)
```
